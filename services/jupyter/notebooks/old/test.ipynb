```python
from pathlib import Path

root_dir = Path(Path.cwd().parent.parent.parent.parent)
csv_file = (root_dir / "datasets/raw_data/cars/cars_sales.csv").as_posix()
parquet_file = (root_dir / "datasets/raw_data/cars/cars_sales2.parquet").as_posix()
```

```python
import subprocess

command = f"clickhouse local --max_memory_usage=2G --max_threads=1 --output_format_parquet_compression=zstd -q \"SELECT * FROM file('{csv_file}', 'CSVWithNames') INTO OUTFILE '{parquet_file}' FORMAT Parquet\""

process = subprocess.Popen(
    command, shell=True, stdout=subprocess.PIPE, stderr=subprocess.PIPE, text=True
)

stdout, stderr = process.communicate()

if process.returncode == 0:
    print("Успешно завершено")
    print("Вывод:", stdout)
else:
    print("Ошибка:", stderr)
```
