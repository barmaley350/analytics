```python
%%time

file = Path(Path.cwd().parent) / "db/raw_data/cars_sales.csv"
chunk_stats = []
for chunk in pd.read_csv(file, chunksize=10000):
    # Вычисляем статистику для чанка
    chunk_desc = chunk.describe()
    chunk_stats.append(chunk_desc)


# Объединяем статистику чанков (упрощённо)
final_stats = pd.concat(chunk_stats).groupby(level=0).mean()
final_stats
```
